```python
import os
import numpy as np
import matplotlib.pyplot as plt

from stable_baselines3 import PPO

from week5_trading_env_v3 import TradingEnvV3


def train_once(rho):

    env = TradingEnvV3(
        rho=rho
    )

    model = PPO(

        "MlpPolicy",

        env,

        learning_rate=3e-4,

        n_steps=2048,

        batch_size=64,

        gamma=0.99,

        verbose=0

    )

    model.learn(
        total_timesteps=30000
    )

    rewards = []

    for episode in range(5):

        obs, info = env.reset()

        done = False

        total_reward = 0.0

        while not done:

            action, _ = model.predict(

                obs,

                deterministic=True

            )

            obs, reward, terminated, truncated, info = env.step(action)

            total_reward += reward

            done = terminated or truncated

        rewards.append(
            total_reward
        )

    env.close()

    return np.mean(
        rewards
    )


def run_sweep():

    rho_values = [

        -0.8,

        -0.4,

        0.0,

        0.4,

        0.8

    ]

    avg_rewards = []

    for rho in rho_values:

        print()

        print(
            f"Training for rho = {rho}"
        )

        mean_reward = train_once(
            rho
        )

        avg_rewards.append(
            mean_reward
        )

        print(
            f"Average Reward = {mean_reward:.2f}"
        )

    return rho_values, avg_rewards


def plot_results(

    rho_values,

    avg_rewards

):

    os.makedirs(

        "plots",

        exist_ok=True

    )

    plt.figure(

        figsize=(8,4)

    )

    plt.plot(

        rho_values,

        avg_rewards,

        marker="o"

    )

    plt.xlabel(

        "Momentum (rho)"

    )

    plt.ylabel(

        "Average Reward"

    )

    plt.title(

        "Effect of Market Momentum"

    )

    plt.grid(True)

    plt.tight_layout()

    plt.savefig(

        "plots/week5_rho_sweep.png"

    )

    plt.close()

    print()

    print(
        "Saved plot to plots/week5_rho_sweep.png"
    )


if __name__ == "__main__":

    rho_values, avg_rewards = run_sweep()

    plot_results(

        rho_values,

        avg_rewards

    )

    print()

    print("Summary")

    for rho, reward in zip(

        rho_values,

        avg_rewards

    ):

        print(

            f"rho = {rho:4.1f}   average reward = {reward:7.2f}"

        )
```
