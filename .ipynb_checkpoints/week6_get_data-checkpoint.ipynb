```python
import os
import yfinance as yf
import pandas as pd

os.makedirs("data", exist_ok=True)

data = yf.download(
    "SPY",
    start="2018-01-01",
    end="2024-01-01",
    auto_adjust=True,
    progress=False
)

data = data[["Close"]]

data["Return"] = data["Close"].pct_change()

data = data.dropna()

data = data.reset_index()

data = data[["Date", "Close", "Return"]]

data.columns = [
    "Date",
    "Close",
    "Return"
]

data.to_csv(
    "data/spy_data.csv",
    index=False
)

print(data.head())

print()

print("Rows:", len(data))

print("Saved -> data/spy_data.csv")
```
