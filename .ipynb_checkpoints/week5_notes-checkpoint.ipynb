# Week 5 Notes

## 1. What changed in TradingEnvV3?

TradingEnvV3 extends the previous trading environment by introducing momentum in the market. Instead of generating completely random daily returns, each return is influenced by the previous return using an autoregressive process controlled by rho. This makes the market behave more like real financial markets where trends can persist for some time.

---

## 2. What does rho represent?

The parameter rho controls the amount of momentum in the market.

- rho close to 1 means strong positive momentum.
- rho close to 0 means almost random returns.
- rho less than 0 creates mean-reverting behaviour.

A larger positive rho makes it easier for an RL agent to identify trends.

---

## 3. Why did we check the autocorrelation?

The environment was designed to generate correlated returns. Computing the lag-1 autocorrelation helped verify that the generated data actually followed the expected momentum.

The observed autocorrelation was close to the selected rho value, so the environment behaved as intended.

---

## 4. PPO Performance

The PPO agent gradually learned to keep positions that matched the market trend.

Compared to a random policy, the trained policy produced higher average rewards and more stable cumulative returns.

Training also became smoother after several thousand timesteps.

---

## 5. Random Agent

The random agent selected actions without considering market information.

Its rewards fluctuated around zero and changed significantly between episodes.

This shows that random trading does not consistently make profits.

---

## 6. Effect of rho

Increasing rho generally improved the PPO agent's performance because stronger momentum created a more predictable environment.

When rho was close to zero, learning became more difficult because price movements were almost random.

Negative rho produced a mean-reverting market which required a different trading behaviour.

---

## 7. What I learned

This week helped me understand how changing the environment affects reinforcement learning.

I learned that even a small change in the market dynamics can make the learning problem easier or harder.

I also learned how PPO can exploit patterns when they exist but struggles when the environment becomes completely random.
