```python
import numpy as np
import matplotlib.pyplot as plt
import os

def moving_average(
    x,
    window=10
):

    if len(x) < window:

        return np.array(x)

    return np.convolve(
        x,
        np.ones(window) / window,
        mode="valid"
    )


reinforce_returns = np.load(
    "reinforce_returns.npy"
)

ppo_returns = np.load(
    "ppo_returns.npy"
)

os.makedirs(
    "plots",
    exist_ok=True
)

plt.figure(
    figsize=(9,4)
)

plt.plot(
    reinforce_returns,
    alpha=0.2
)

plt.plot(
    moving_average(
        reinforce_returns
    ),
    label="REINFORCE"
)

plt.plot(
    ppo_returns,
    alpha=0.2
)

plt.plot(
    moving_average(
        ppo_returns
    ),
    label="PPO"
)

plt.xlabel("Episode")

plt.ylabel("Return")

plt.title(
    "REINFORCE vs PPO on CartPole-v1"
)

plt.legend()

plt.tight_layout()

plt.savefig(
    "plots/reinforce_vs_ppo.png"
)

plt.show()
```
