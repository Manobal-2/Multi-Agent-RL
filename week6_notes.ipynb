# Week 6 Notes

## 1. Objective

This week focused on moving from a synthetic trading environment to a historical market dataset. Instead of generating random price movements, the environment used actual SPY daily returns downloaded from Yahoo Finance. This made the reinforcement learning problem more realistic.

---

## 2. Data Collection

Historical data was downloaded using the yfinance library.

The adjusted closing prices were used to calculate daily percentage returns. These returns became the observations used by the trading environment.

Using historical data allows the agent to learn from real market behaviour instead of simulated randomness.

---

## 3. Historical Trading Environment

The environment was similar to the previous weeks but replaced simulated returns with historical returns.

The observation contained:

- Previous price returns
- Current trading position

The available actions were:

- Long
- Hold
- Short

The reward depended on the current position, daily market return and transaction cost.

---

## 4. PPO Training

The PPO agent was trained using Stable-Baselines3.

Training gradually improved the cumulative reward as the agent experienced more historical market data.

The moving average of rewards became smoother over time, indicating more stable learning.

---

## 5. Evaluation

After training, the PPO agent was evaluated on multiple episodes.

The cumulative reward generally outperformed the random trading strategy.

Although some episodes still produced losses, the trained policy was much more consistent than taking random actions.

---

## 6. Random Agent Comparison

The random policy selected actions without considering market history.

Because the decisions were completely random, the rewards fluctuated significantly between episodes.

This demonstrated that profitable trading requires learning useful patterns rather than making random decisions.

---

## 7. What I Learned

This week helped connect reinforcement learning with real financial data.

I learned that replacing simulated environments with historical data introduces additional uncertainty and makes the learning problem more challenging.

I also understood how transaction costs reduce profits and why realistic environments are important when evaluating trading strategies.

---

## 8. Challenges

The biggest challenge was preparing the historical data correctly before training.

Another challenge was understanding how PPO interacts with a sequential financial dataset instead of a fully simulated environment.

I also learned that RL performance can vary depending on the quality of the available market data.

---

## 9. Future Improvements

Some improvements that could be explored are:

- Add technical indicators such as RSI and moving averages.
- Include trading volume as an observation.
- Use multiple stocks instead of only SPY.
- Train on one time period and evaluate on a completely different period.
- Compare PPO with other reinforcement learning algorithms.
